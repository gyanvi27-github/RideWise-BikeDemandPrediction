```python
import seaborn as sns

# Set style
sns.set_theme(style="whitegrid")

# Load dataset (assuming you already read it earlier)
df_day = df_day.copy()   #  convert df to df_day

# Convert date column to datetime
df_day['dteday'] = pd.to_datetime(df_day['dteday'])

# Aggregate by month for clearer trend
monthly = df_day.groupby(df_day['dteday'].dt.to_period("M"))['cnt'].sum().reset_index()
monthly['dteday'] = monthly['dteday'].dt.to_timestamp()

# Add rolling average (3-month window)
monthly['rolling_avg'] = monthly['cnt'].rolling(window=3).mean()

# Plot
plt.figure(figsize=(14,7))
sns.lineplot(x="dteday", y="cnt", data=monthly, label="Monthly Rentals", marker="o")
sns.lineplot(x="dteday", y="rolling_avg", data=monthly, label="3-Month Rolling Avg", color="red")

# Titles & labels
plt.title(" Bike Rentals Trend with Seasonality & Rolling Average", fontsize=16, weight="bold")
plt.xlabel("Month", fontsize=12)
plt.ylabel("Total Rentals", fontsize=12)

# Annotate peak
peak = monthly.loc[monthly['cnt'].idxmax()]
plt.annotate(f"Peak: {peak['cnt']:,}", 
             xy=(peak['dteday'], peak['cnt']), 
             xytext=(peak['dteday'], peak['cnt']+2000),
             arrowprops=dict(facecolor='black', shrink=0.05),
             fontsize=10, color="darkblue")

# Rotate ticks
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x="season", y="cnt", data=df_day, estimator=sum)
plt.title("Total Rentals by Season")
plt.xlabel("Season (1:Winter, 2:Spring, 3:Summer, 4:Fall)")
plt.ylabel("Total Rentals")
plt.show()

# 3. Stacked Bar Chart: Casual vs Registered Users per Month
monthly = df_day.groupby("mnth")[["casual","registered"]].sum()
monthly.plot(kind="bar", stacked=True, figsize=(10,6))
plt.title("Casual vs Registered Rentals per Month")
plt.xlabel("Month")
plt.ylabel("Rentals")
plt.show()
```

```python
# Map weekday numbers to names #4.box plot
weekday_map = {0:"Sunday",1:"Monday",2:"Tuesday",3:"Wednesday",4:"Thursday",5:"Friday",6:"Saturday"}
df_day["weekday_name"] = df_day["weekday"].map(weekday_map)

plt.figure(figsize=(10,6))
sns.boxplot(x="weekday_name", y="cnt", data=df_day, palette="Set2")

# Add overall mean line
plt.axhline(df_day["cnt"].mean(), color="red", linestyle="--", label=f"Overall Mean: {df_day['cnt'].mean():.0f}")

plt.title("Rental Variation Across Weekdays", fontsize=14, weight="bold")
plt.xlabel("Weekday")
plt.ylabel("Total Rentals")
plt.legend()
plt.show()
```

```python
# Map season codes to names
import seaborn as sns
season_map = {1:"Winter", 2:"Spring", 3:"Summer", 4:"Fall"}
df_day["season_name"] = df_day["season"].map(season_map)

plt.figure(figsize=(12,7))

# Scatter plot colored by season
sns.scatterplot(x="temp", y="cnt", data=df_day, hue="season_name",
                palette="Set2", alpha=0.6)

# Overall regression line
sns.regplot(x="temp", y="cnt", data=df_day, scatter=False, color="black",
            line_kws={"label":"Overall Trend", "linewidth":2})

# Add regression lines per season
for season, group in df_day.groupby("season_name"):
    sns.regplot(x="temp", y="cnt", data=group, scatter=False, ci=None,
                label=f"{season} Trend")

# Titles & labels
plt.title(" Bike Rentals vs Temperature\nStrategic Comfort Zone & Seasonal Demand", 
          fontsize=16, weight="bold")
plt.xlabel("Normalized Temperature (0 = cold, 1 = hot)", fontsize=12)
plt.ylabel("Total Rentals", fontsize=12)

# Comfort zone annotation
plt.axvspan(0.4, 0.7, color="yellow", alpha=0.2)
plt.text(0.55, df_day["cnt"].max()*0.9, "Comfort Zone\n(High Rentals)",
         fontsize=11, color="darkblue", ha="center")

# Demand bands (low, medium, high rentals)
max_cnt = df_day["cnt"].max()
plt.axhspan(0, max_cnt*0.3, color="red", alpha=0.05)
plt.axhspan(max_cnt*0.3, max_cnt*0.6, color="orange", alpha=0.05)
plt.axhspan(max_cnt*0.6, max_cnt, color="green", alpha=0.05)
plt.text(0.05, max_cnt*0.15, "Low Demand", fontsize=10, color="red")
plt.text(0.05, max_cnt*0.45, "Medium Demand", fontsize=10, color="orange")
plt.text(0.05, max_cnt*0.75, "High Demand", fontsize=10, color="green")

# Horizontal mean lines per season
for season, group in df_day.groupby("season_name"):
    mean_val = group["cnt"].mean()
    plt.axhline(mean_val, linestyle="--", alpha=0.4,
                label=f"{season} Avg: {int(mean_val)}")

# Legend enriched with KPIs
plt.legend(title="Seasonal Trends & KPIs", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
corr = df_day[["temp","atemp","hum","windspeed","casual","registered","cnt"]].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Between Weather & Rentals")
plt.show()
```

```python
sns.set_theme(style="whitegrid")
 
# 7. Rentals by Season (Pie Chart + Avg Rentals)

# Map season codes to names
season_map = {1:"Winter", 2:"Spring", 3:"Summer", 4:"Fall"}
df_day["season_name"] = df_day["season"].map(season_map)

# Total rentals by season
season_totals = df_day.groupby("season_name")["cnt"].sum()

# Average rentals by season
season_avg = df_day.groupby("season_name")["cnt"].mean()
# Pie chart
plt.figure(figsize=(8,8))
plt.pie(season_totals, labels=season_totals.index, autopct="%1.1f%%", 
        startangle=90, colors=sns.color_palette("Set2"))
plt.title(" Share of Rentals by Season", fontsize=14, weight="bold")
plt.show()

# Bar chart for average rentals
plt.figure(figsize=(8,6))
sns.barplot(x=season_avg.index, y=season_avg.values, palette="Set2")
plt.title("Average Rentals per Season", fontsize=14, weight="bold")
plt.ylabel("Average Rentals")

# Highlight min and max
min_season = season_avg.idxmin()
max_season = season_avg.idxmax()
plt.text(season_avg.index.tolist().index(min_season), season_avg.min()+200, 
         f"Lowest: {int(season_avg.min())}", color="red", ha="center")
plt.text(season_avg.index.tolist().index(max_season), season_avg.max()+200, 
         f"Highest: {int(season_avg.max())}", color="green", ha="center")
plt.show()
# Map weather codes to names
weather_map = {1:"Clear", 2:"Mist", 3:"Light Snow/Rain", 4:"Heavy Rain"}
df_day["weather_name"] = df_day["weathersit"].map(weather_map)

# Average rentals by weather
weather_avg = df_day.groupby("weather_name")["cnt"].mean().sort_values(ascending=False)

plt.figure(figsize=(8,6))
sns.barplot(x=weather_avg.index, y=weather_avg.values, palette="coolwarm")
plt.title(" Impact of Weather on Bike Rentals", fontsize=14, weight="bold")
plt.ylabel("Average Rentals")

# Annotate best and worst
plt.text(weather_avg.index.tolist().index(weather_avg.idxmax()), weather_avg.max()+200, 
         f"Best: {int(weather_avg.max())}", color="green", ha="center")
plt.text(weather_avg.index.tolist().index(weather_avg.idxmin()), weather_avg.min()+200, 
         f"Worst: {int(weather_avg.min())}", color="red", ha="center")
plt.show()

```

```python
 
plt.figure(figsize=(10,6))

# Histogram with KDE curve
sns.histplot(df_day["cnt"], bins=30, kde=True, color="skyblue")

# Add mean & median lines
mean_val = df_day["cnt"].mean()
median_val = df_day["cnt"].median()

plt.axvline(mean_val, color="red", linestyle="--", linewidth=2, label=f"Mean: {mean_val:.0f}")
plt.axvline(median_val, color="green", linestyle="--", linewidth=2, label=f"Median: {median_val:.0f}")

# Annotate most common range
mode_range = df_day["cnt"].mode()[0]
plt.annotate(f"Most common rentals  {mode_range}", 
             xy=(mode_range, 50), xytext=(mode_range+1000, 100),
             arrowprops=dict(facecolor='black', shrink=0.05),
             fontsize=10, color="darkblue")

# Titles & labels
plt.title("Distribution of Daily Bike Rentals", fontsize=15, weight="bold")
plt.xlabel("Daily Rental Count", fontsize=12)
plt.ylabel("Frequency (Number of Days)", fontsize=12)

plt.legend()
plt.tight_layout()
plt.show()
```
